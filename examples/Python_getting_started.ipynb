# Simple Python examples for SYNERGY dataset

Start with importing `Dataset` and `iter_datasets`.

```python
from pprint import pprint

from synergy_dataset import Dataset, iter_datasets
```

## Get each dataset as pandas DataFrame

```python
for d in iter_datasets():

    df = d.to_frame()

    print(d.name, type(df))
```

```python
df.head()
```

## Get each dataset as dict

```python
for d in iter_datasets():

    obj = d.to_dict()

    print(d.name, type(obj))
```

```python
pprint(list(obj.values())[0])
```

## Get single dataset

```python
d = Dataset("Hall_2012")

d.to_frame()
```

## Get other variables than title and abstract

```python
d = Dataset("Hall_2012")

obj = d.to_dict(["title", "referenced_works"])

pprint(list(obj.values())[0])
```

## Get metadata about the final publication

```python
d = Dataset("Hall_2012")

# the full OpenAlex work
d.metadata
```

## Build one big dataset

```python
import pandas as pd

pd.concat([d.to_frame() for d in iter_datasets()], axis=0)
```

## Get work identifiers

https://docs.openalex.org/api-entities/works/work-object#ids

```python
d = Dataset("Appenzeller-Herzog_2019")
ids = d.to_dict(["ids"])

# output the first 10 ids
from itertools import islice

dict(islice(ids.items(), 10))
```
